# Measure of Central Tendency

Measures of central tendency are statistical measures that provide information about the central or typical value of a dataset. They give us a sense of where the data is centered or concentrated. The three common measures of central tendency are: 

- Mean
- Median
- Mode

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Generate random numbers from a skewed normal distribution
x = np.random.standard_t(10, size=1000)
x[:15]
```

In this approach, we make use of the `np.random.standard_t` function from the `NumPy` library to generate random numbers from a t-distribution. By setting the degrees of freedom parameter (df) to a high value (e.g., 10), the t-distribution approximates a skewed normal distribution.

In the example `np.random.standard_t(10, size=1000)`, the number `10` represents the degrees of freedom parameter (`df`).

The t-distribution is defined by its degrees of freedom, which determines the shape of the distribution. In the context of the t-distribution, the degrees of freedom represent the number of independent observations used to estimate a population parameter.

By setting the degrees of freedom to `10` in this example, we are generating random numbers from a t-distribution with 10 degrees of freedom, which approximates a skewed normal distribution. You can adjust the degrees of freedom parameter based on your specific requirements or desired shape of the distribution.

The resulting `x` will be an array of 1000 random numbers following a skewed normal distribution. You can adjust the parameters as needed to customize the shape and characteristics of the skewed distribution.

```python
_ = plt.hist(x, color = 'blue')
```

## Mean

The mean, denoted by $\mu$, is a measure of central tendency that represents the average value of a dataset. It is calculated by summing all the values in the dataset and dividing by the number of observations.

\begin{equation}
\mu = \frac{\sum_{i=1}^{n} x_i}{n}
\end{equation}

```python
mean = np.mean(x)
print("Mean:", mean)
```

```python
plt.hist(x, color = 'blue')
plt.axvline(x = mean, color='yellow')
```

## Median

The median is another measure of central tendency that represents the middle value of a sorted dataset. If the dataset has an odd number of observations, the median is the middle value. If the dataset has an even number of observations, the median is the average of the two middle values.

We can calculate the median values in two different ways:

### Odd Number of Observations:

If the dataset has an odd number of observations (n), the median is the middle value when the dataset is arranged in ascending order. The formula for calculating the median in this case is:

\begin{equation}
\text{Median} = x_{\left(\frac{n+1}{2}\right)}
\end{equation}

### Even Number of Observations:

If the dataset has an even number of observations (n), the median is the average of the two middle values when the dataset is arranged in ascending order. The formula for calculating the median in this case is:

\begin{equation}
\text{Median} = \frac{x_{\left(\frac{n}{2}\right)} + x_{\left(\frac{n}{2}+1\right)}}{2}
\end{equation}

```python
median = np.median(x)
print("Median:", median)
```

```python
plt.hist(x, color = 'blue')
plt.axvline(x = median, color='yellow')
```

## Mode

The mode is a measure of central tendency that represents the most frequently occurring value(s) in a dataset. A dataset can have one mode (unimodal) or multiple modes (multimodal).

```python
mode = stats.mode(x).mode
print("Mode:", mode)
```

```python
plt.hist(x, color = 'blue')
plt.axvline(x = mode, color='yellow')
```

# Measure of Dispersion/Variability

A measure of dispersion, also known as a measure of variability or spread, is a statistical metric that quantifies how the data points in a dataset are spread out or dispersed around a central value. It provides insight into the distribution and variability of the data.

Measures of dispersion help us understand the range or extent to which the data points deviate from a central tendency. They provide additional information beyond the measures of central tendency (such as mean, median, and mode) by capturing the spread, scatter, or variability within the dataset.

Some of the common measures of dispersion are:

- Variance
- Standard Deviation
- Range
- Interquartile Range (IQR)
- Mean Absolute Deviation (MAD)

## Variance

The variance, denoted by $\sigma^2$, is another measure of the spread of a dataset. It is the average of the squared deviations from the mean. Variance provides a measure of how the individual data points differ from the mean.

\begin{equation}
\sigma^2 = \frac{\sum_{i=1}^{n} (x_i - \mu)^2}{n}
\end{equation}

```python
variance = np.var(x)
print("Variance:", variance)
```

```python
plt.hist(x, color = 'blue')
plt.axvline(x = mean, color='yellow')
plt.axvline(x = variance, color='red')
```

## Standard Deviation

The standard deviation, denoted by $\sigma$, measures the dispersion or spread of a dataset. It quantifies how much the values deviate from the mean. A higher standard deviation indicates greater variability in the data.

\begin{equation}
\sigma = \sqrt{\frac{\sum_{i=1}^{n} (x_i - \mu)^2}{n}}
\end{equation}

```python
std = np.std(x)
print("Standard Deviation:", std)
```

```python
plt.hist(x, color = 'blue')
plt.axvline(x = std, color='red')
plt.axvline(x = mean, color='yellow')
```

## Range

The range is a measure of dispersion that indicates the difference between the maximum and minimum values in a dataset. It provides a simple representation of the spread of the data.

\begin{equation}
Range = Maximum\ value\ -\ Minimum\ value
\end{equation}

```python
range = np.max(x) - np.min(x)
print("Range:", range)
```

## Interquartile Range (IQR)

The interquartile range is a measure of dispersion that captures the spread of the middle 50% of the data, excluding the extreme values. It is calculated as the difference between the upper quartile (Q3) and the lower quartile (Q1).

\begin{equation}
IQR = Q3 - Q1
\end{equation}

```python
q1 = np.percentile(x, 25)
q3 = np.percentile(x, 75)

iqr = q3 - q1
print("IQR:", iqr)
```

## Mean Absolute Deviation (MAD)

The mean absolute deviation is a measure of dispersion that quantifies the average absolute difference between each data point and the mean. It represents the average distance of each data point from the central value.

\begin{equation}
MAD = \frac{1}{n} \sum_{i=1}^{n} |x_i - \text{mean}|
\end{equation}

```python
mad = np.mean(np.abs(x - mean))
print("MAD:", mad)
```

```python
plt.hist(x, color = 'blue')
plt.axvline(x = mad, color='red')
plt.axvline(x = mean, color='yellow')
```
