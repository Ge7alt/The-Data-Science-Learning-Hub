# Central Limit Theorem

The Central Limit Theorem is a fundamental concept in statistics that describes the behavior of the sum or average of a large number of independent and identically distributed (i.i.d.) random variables. It states that regardless of the shape of the original population distribution, the sampling distribution of the sum or average tends to follow a normal distribution as the sample size increases.

The CLT has immense practical significance since it allows us to make statistical inferences about population parameters even when we don't know the underlying distribution. It forms the basis for many statistical techniques and hypothesis tests.

## Assumptions of the Central Limit Theorem
Before we dive into the details, let's outline the assumptions of the Central Limit Theorem:

- **Independence:** The random variables being sampled must be independent of each other.
- **Identical Distribution:** The random variables must have the same distribution, with the same mean and standard deviation.
- **Sample Size:** The sample size should be sufficiently large. The CLT usually works well for sample sizes greater than 30, although there are cases where it can still provide reasonably accurate results for smaller sample sizes.

## Statement of the Central Limit Theorem
The Central Limit Theorem can be stated as follows:

Let $X_1, X_2, \ldots, X_n$ be a sequence of $n$ independent and identically distributed random variables with mean $\mu$ and standard deviation $\sigma$. Consider the random variable $Y_n = \frac{X_1 + X_2 + \ldots + X_n}{\sqrt{n}}$. Then, as $n$ approaches infinity, the distribution of $Y_n$ approaches a standard normal distribution with mean $0$ and standard deviation $1$.


In simpler terms, the sum or average of a large number of independent and identically distributed random variables follows an approximately normal distribution.

## Demonstrating the Central Limit Theorem

Let's demonstrate the Central Limit Theorem using Python and `matplotlib`. We'll generate samples from different distributions and show how the distribution of the sample means converges to a normal distribution as the sample size increases.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Uniform Distribution

Let's start with a simple example where the underlying population distribution follows a uniform distribution.

```python
# Set the seed for reproducibility
np.random.seed(42)

# Generate samples from a uniform distribution (0, 1)
population = np.random.uniform(0, 1, size=100000)

# Plot the population distribution
plt.figure(figsize=(8, 6))
plt.hist(population, bins=50, density=True, color='blue', alpha=0.6)
plt.title('Population Distribution (Uniform)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

The above code generates 100,000 samples from a uniform distribution between 0 and 1 and plots the population distribution.

Now, let's generate samples of different sizes from the population and calculate the sample means:

```python
# Sample sizes to consider
sample_sizes = [5, 10, 30, 100, 1000]

# Generate samples of different sizes and calculate sample means
sample_means = []
for n in sample_sizes:
    samples = np.random.choice(population, size=(1000, n), replace=True)
    means = np.mean(samples, axis=1)
    sample_means.append(means)

# Plot the distribution of sample means for each sample size
plt.figure(figsize=(12, 8))
for i, means in enumerate(sample_means):
    plt.subplot(2, 3, i+1)
    plt.hist(means, bins=30, density=True, color='green', alpha=0.6)
    plt.title(f'Sample Size: {sample_sizes[i]}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

In the code above, we generate samples of different sizes (5, 10, 30, 100, 1000) from the population and calculate the sample means. Then, we plot the distribution of sample means for each sample size.

By running the code, you should see a figure with six subplots. As the sample size increases, the distribution of the sample means approaches a bell-shaped curve, indicating the convergence to a normal distribution.

### Exponential Distribution

Now, let's demonstrate the Central Limit Theorem using a different population distribution, such as the exponential distribution.

```python
# Set the seed for reproducibility
np.random.seed(42)

# Generate samples from an exponential distribution
population = np.random.exponential(scale=1, size=100000)

# Plot the population distribution
plt.figure(figsize=(8, 6))
plt.hist(population, bins=50, density=True, color='blue', alpha=0.6)
plt.title('Population Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

This code generates 100,000 samples from an exponential distribution with a scale parameter of 1 and plots the population distribution.

Next, we'll generate samples of different sizes and calculate the sample means, similar to the previous example:

```python
# Sample sizes to consider
sample_sizes = [5, 10, 30, 100, 1000]

# Generate samples of different sizes and calculate sample means
sample_means = []
for n in sample_sizes:
    samples = np.random.choice(population, size=(1000, n), replace=True)
    means = np.mean(samples, axis=1)
    sample_means.append(means)

# Plot the distribution of sample means for each sample size
plt.figure(figsize=(12, 8))
for i, means in enumerate(sample_means):
    plt.subplot(2, 3, i+1)
    plt.hist(means, bins=30, density=True, color='green', alpha=0.6)
    plt.title(f'Sample Size: {sample_sizes[i]}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

By running the code, you'll observe the distribution of sample means approaching a normal distribution as the sample size increases, even though the population distribution is not normal.
