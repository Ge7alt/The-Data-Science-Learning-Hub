# Random Variables

A random variable is a variable that takes on different numerical values depending on the outcome of a random event. In probability theory, random variables are used to model the uncertainty associated with a particular event or experiment.

Properties of random variables include:

- Domain: the set of all possible values that the random variable can take on.
- Probability distribution: the probability of each possible value of the random variable.
- Mean: the expected value of the random variable, which is the weighted average of all possible values weighted by their probabilities.
- Variance: a measure of how much the random variable varies from its mean value.
- Probability density function: for continuous random variables, a function that describes the relative likelihood of each possible value within a given range.
- Cumulative distribution function: a function that gives the probability that a random variable takes a value less than or equal to a certain value.

These properties are important for understanding the behavior of random variables and their applications in probability theory and statistics.

## Discrete and Continuous Random Variables

Discrete random variables take on a countable number of possible values. For example, the number of heads that occur when flipping a coin five times is a discrete random variable, as it can only take on the values 0, 1, 2, 3, 4, or 5.

Continuous random variables, on the other hand, can take on any value within a certain range. For example, the height of an adult is a continuous random variable, as it can take on any value between a certain minimum and maximum height.

The probability distribution of a random variable describes the likelihood of each of its possible values. For a discrete random variable, this is given by a probability mass function (PMF), while for a continuous random variable, it is given by a probability density function (PDF).

## Probability Mass Function (PMF)

A probability mass function (PMF) is a function that maps each possible value of a discrete random variable to its probability of occurrence. In other words, it gives the probability of each possible outcome of the random variable.

`For example`, consider a fair six-sided die. The possible outcomes are the integers `1` through `6`. The PMF of this random variable is:

\begin{equation}
P(1) = P(2) = P(3) = P(4) = P(5) = P(6) = 1/6
\end{equation}

This means that the probability of rolling a 1, 2, 3, 4, 5, or 6 is each `1/6`.

The PMF has the following properties:

- The PMF is non-negative: `P(x) ≥ 0` for all `x` in the domain of the random variable.
- The sum of the probabilities is `1`: `∑P(x) = 1`, where the sum is taken over all possible values of `x`.
- The PMF can be used to calculate the probability of any event involving the random variable.

In Python, the PMF of a discrete random variable can be implemented using a dictionary, where the keys are the possible values of the random variable and the values are their corresponding probabilities. For example, the PMF of a fair six-sided die can be implemented as follows:

```python
pmf = {1: 1/6, 2: 1/6, 3: 1/6, 4: 1/6, 5: 1/6, 6: 1/6}
```

## Probability Density Function (PDF)

Probability density function (PDF) is a function that describes the relative likelihood for a continuous random variable to take on a given value. The PDF is often denoted by f(x), where x is the variable, and it is defined such that the probability of x being between a and b is given by the integral of f(x) between a and b:

\begin{equation}
P(a ≤ x ≤ b) = ∫[a,b] f(x)dx
\end{equation}

An example of a continuous random variable with a PDF is the normal distribution, which was discussed earlier.

### Properties of PDF

- The area under the PDF curve is always equal to `1`.
- The PDF curve is non-negative for all values of `x`.
- The probability of x being equal to a specific value is always zero (since the area under a single point is zero).
- The expected value of a continuous random variable can be calculated as the integral of `x` times the PDF over all possible values of `x`.
- The variance of a continuous random variable can be calculated as the integral of `(x-E(x))^2` times the PDF over all possible values of `x`.

### Moments of PDF

The moments of a probability density function describe the properties of the distribution. The moments of a continuous random variable X with probability density function f(x) are given by:

- The nth moment of `X` is defined as `E(X^n) = ∫x^n f(x) dx` over the entire range of `X`.
- The expected value of `X` is the first moment `E(X) = ∫x f(x) dx`.
- The variance of `X` is the second central moment `Var(X) = E((X - E(X))^2) = E(X^2) - (E(X))^2`.
- The standard deviation of `X` is the square root of the variance, `σ = sqrt(Var(X))`.
- Higher moments can also be defined, but the first two moments are the most commonly used.

Moments of a probability density function can provide insight into the shape, spread, and skewness of the distribution. For example, the variance measures the spread of the distribution, while the skewness measures its asymmetry.

The moments of a probability density function can also be used to estimate the parameters of the distribution, such as the mean and standard deviation.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# define the parameters of the normal distribution
mu = 0
sigma = 1

# create an array of x values
x = np.linspace(-5, 5, 100)

# calculate the corresponding y values using the normal probability density function
y = norm.pdf(x, mu, sigma)

# plot the normal probability density function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Probability density')
plt.title('Normal probability density function')
plt.show()
```

This code creates an array of `x` values ranging from `-5` to `5`, calculates the corresponding `y` values using the probability density function of a normal distribution with mean `mu=0` and standard deviation `sigma=1`, and then plots the result using `matplotlib`. The resulting plot is a bell-shaped curve centered at `x=0`.

## Cumulative Distribution Function

The cumulative distribution function (CDF) is a function that maps the probability of a random variable being less than or equal to a certain value. For a discrete random variable, the CDF is the sum of the probability mass function (PMF) up to a certain value. For a continuous random variable, the CDF is the integral of the probability density function (PDF) up to a certain value.

The CDF can be used to calculate the probability of a random variable taking a value within a certain range, as well as to calculate quantiles and percentiles.

In general, the CDF of a random variable `X` is defined as:

\begin{equation}
F(x) = P(X ≤ x)
\end{equation}

where `F(x)` is the CDF of `X`, and `P(X ≤ x)` is the probability that `X` is less than or equal to `x`.

The CDF has the following properties:

- It is non-decreasing: `F(x1) ≤ F(x2)` if `x1 ≤ x2`.
- It is right-continuous: `lim F(x)` as `x → a+ = F(a)` for any `a`.
- It is bounded: `0 ≤ F(x) ≤ 1` for all `x`.
- It approaches `0` as `x → -∞` and approaches `1` as `x → +∞`.
- The probability of `X` taking a value between `x1` and `x2` is given by `F(x2) - F(x1)`.

The inverse of the CDF, called the quantile function or percentile function, gives the value `x` such that `F(x) = p`, where `p` is a given probability between `0` and `1`.

The first derivative of the CDF is the PDF.

Example: Suppose we have a continuous random variable `X` that follows a standard normal distribution (i.e., with mean `0` and standard deviation `1`). We can calculate the probability that `X` is less than or equal to `1.5` using the CDF:

\begin{equation}
F(1.5) = P(X ≤ 1.5) = ∫_{-∞}^{1.5} f(x) dx
\end{equation}

where `f(x)` is the PDF of the standard normal distribution, which is given by:

\begin{equation}
f(x) = (1/√(2π)) * e^(-(x^2)/2)
\end{equation}

Using numerical integration or a table of values, we can find that `F(1.5) = 0.9332`.

`For example:` to calculate the CDF of a standard normal distribution at `x = 1.5`, we can use the following code:

```python
import scipy.stats as stats

x = 1.5
cdf = stats.norm.cdf(x)
print(cdf)
```

Similarly, the quantile function (ppf) method can be used to find the inverse of the CDF, i.e., the value `x` such that `F(x) = p`, where `p` is a given probability between 0 and 1. For example, to find the value `x` such that the probability of a standard normal distribution being less than or equal to `x` is 0.95, we can use the following code:

```python
p = 0.95
x = stats.norm.ppf(p)
print(x)
```

Here's the implementation in Python for plotting the CDF and PDF for the normal distribution with mean=170 and standard deviation=5:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set the mean and standard deviation for the normal distribution
mu = 170
sigma = 5

# Create an array of x values ranging from 100 to 240 with a step of 0.1
x = np.arange(100, 240, 0.1)

# Calculate the corresponding PDF and CDF values for the x values
pdf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)

# Create a figure and axis objects
fig, ax = plt.subplots(figsize=(8, 5))

# Plot the PDF and CDF curves
ax.plot(x, pdf, label='PDF')
ax.plot(x, cdf, label='CDF')

# Set axis labels and title
ax.set_xlabel('Height (cm)')
ax.set_ylabel('Probability density')
ax.set_title('Normal Distribution with mean=170, std=5')

# Show legend
ax.legend()

# Show the plot
plt.show()
```
